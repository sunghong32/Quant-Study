```python
import yfinance as yf
```

```python
data = yf.download(
    tickers = "SOXS",
    period = "1y",
    interval = "1d",
    ignore_tz = True,
    prepost = False
)
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
from finta import TA
```

```python
from finta.utils import to_dataframe, resample, trending_down, trending_up
```

```python
for c in data.columns:
    data[c.lower()] = data[c]
    del data[c]
```

```python
olhc = resample(data, "5d")
olhc = data
```

```python
smaData = TA.SMA(olhc, 5)
ssmaData = TA.SSMA(olhc)
emaData = TA.EMA(olhc)
bbandsData = TA.BBANDS(olhc)
rsiData = TA.RSI(olhc)
macdData = TA.MACD(olhc)
obvData = TA.OBV(olhc)
```

```python
plt.figure(figsize=(25,12))
plt.plot(smaData, label="sma")
plt.plot(ssmaData, label="ssma")
plt.plot(emaData, label="ema")
plt.legend()
```

```python
plt.figure(figsize=(25,12))

plt.subplot(3, 1, 1)  
plt.plot(smaData, label="sma")
# plt.plot(bbandsData, label="bbands")
plt.plot(rsiData, label="rsi")
plt.xlabel('period')
plt.legend()

plt.subplot(3,1,2)
plt.plot(obvData, label="obv")
plt.xlabel('period')
plt.legend()

plt.subplot(3,1,3)
plt.plot(macdData, label="macd")
plt.xlabel('period')
plt.legend()

plt.tight_layout()
plt.show()
```
